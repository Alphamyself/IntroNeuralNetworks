```python
import numpy as np
import pandas as pd
```

```python
# Import the titantic data set: raw_data
# Data comes from the Kaggle Data Set. 
# See explaination of the features here: https://www.kaggle.com/c/titanic/data
raw_data = pd.read_csv("../Data/titantic_training_data.csv")

# Print the first 6 rows in raw_data
raw_data.head()
```

```python
# PassengerId, Name and Ticket are not imporant to predictions
relevant_cols = ["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]
raw_data = raw_data[relevant_cols]
raw_data.head()
```

```python
# Now, we need to take care of null values. 
# Raw Data will often have many gaps or nulls.
# See how many null values you have in the data
raw_data.apply(lambda x: sum(pd.isnull(x)))
```

```python
# Age and Embarked have 177 and 2 gaps/NaN cells respectively. 
# Becuase they are differing data types, they should be handled differently.

# Age is a numerical value which has a mean. Fill the NA with the mean.
raw_data['Age'].fillna((raw_data['Age'].mean()), inplace=True)

# Embarked is a catagorical value which has a mode. Fill the NA with the mode.
raw_data['Embarked'].fillna((raw_data['Embarked'].mode()[0]), inplace=True)

# Make sure there are no more gaps
raw_data.apply(lambda x: sum(pd.isnull(x)))
```

```python
# Next, we need to see which variables are catagorical in nature.
# We can do this by seeing how many unique values there are for each variable.
raw_data.apply(lambda x: len(x.unique()))
```

```python
# Looks like Survived, Embarked, Pclass, and Sex are certainly catagories/factors
# (to convince yourself, look at the feature descriptions from the link above)
# Encode the catagorical variables since Neural Networks only work with numerical values
# This is called 'one hot encoding' (One of several ways to deal with catagorical variables)
raw_data = pd.get_dummies(raw_data, columns=["Sex", "Pclass", "Embarked"])
raw_data.head()
```

```python
from keras.utils import to_categorical

# As a last step, seperate the target values (Survived) from the predictor data
targets_single_col = raw_data.Survived

# Like we did before with pandas, get the dummy/one hot encoding for our target Survived
targets = to_categorical(targets_single_col)

# Set predictors as the rest of the raw data without 'Survived'
predictors_df = raw_data.drop("Survived", axis=1)

# Convert the raw_data pandas dataframe to a numpy matrix for input to Keras
predictors = predictors_df.as_matrix()

# Print one hot encoded Survived targets
print("Targets\n", targets)

print("Targets and predictors from titantic data set loaded successfully.")
```
